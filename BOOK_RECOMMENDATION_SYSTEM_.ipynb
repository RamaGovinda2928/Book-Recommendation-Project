Tried to compute cosine similarity but facing issues with RAM.so tried to compute cosine similarity using sparse matrix..but same issues arising.So using approximate similarity methods such as locality-sensitive hashing (LSH) or MinHash. These methods trade off some accuracy for lower memory usage and faster computation.

MinHash, a technique for estimating Jaccard similarity between sets, to approximate similarity between items in a large dataset with reduced memory usage

In this code:

We use the CountVectorizer to convert the text features (book titles and authors) into a matrix representation.
We initialize a MinHashLSHForest object with a specified number of permutations.
We create MinHash objects for each item in the dataset and add them to the forest.
We index the MinHash objects in the forest.
Finally, we query the forest for similar items to a given item (in this case, item 0) and print the results.
The MinHashLSHForest allows us to approximate similarity between items in a large dataset with reduced memory usage and faster computation compared to exact methods like cosine similarity. Adjust the number of permutations (num_perm) based on your requirements for accuracy and memory usage.

```python
import pandas as pd
from datasketch import MinHash, MinHashLSHForest
# Initialize MinHashLSHForest
forest = MinHashLSHForest(num_perm=256)

# Create MinHash objects for each item (book title and author combination)
minhash_list = []
for index, row in data.iterrows():
    minhash = MinHash(num_perm=256)
    for word in (row['book_title'] + " " + row['book_author']).split():
        minhash.update(word.encode('utf-8'))
    minhash_list.append(minhash)
    forest.add(str(index), minhash)

# Index the MinHash objects
forest.index()
```

```python
# Query for similar items
query_title = "Classical Mythology"  # Title of the book you want to find similar books for
query_idx = data.index[data['book_title'] == query_title][0]  # Get the index corresponding to the query title
query_minhash = minhash_list[query_idx]
result = forest.query(query_minhash, 2)  # Increase the result size to 4 to account for filtering out the query book

# Print similar book titles
print("Similar books to:", query_title)
for idx in result:
    if int(idx) != query_idx:  # Filter out the query book
        print("-", data.loc[int(idx), 'book_title'])
```

## Deployment

```python
from surprise import Reader, Dataset, SVD
from surprise.model_selection import train_test_split
import pandas as pd


reader = Reader(rating_scale=(1, 10))
data = Dataset.load_from_df(df_all[['user_id', 'book_title', 'book_rating']], reader)


trainset, testset = train_test_split(data, test_size=0.2, random_state=42)

model = SVD()
model.fit(trainset)

def get_recommendations(user_id, n, model, all_books):
    user_unrated_books = df_all[df_all['user_id'] == user_id]['book_title'].unique()
    all_books = df_all['book_title'].unique()
    to_predict = [book for book in all_books if book not in user_unrated_books]

    test_data = [(user_id, book, 0) for book in to_predict]
    predictions = model.test(test_data)

    top_n = sorted(predictions, key=lambda x: x.est, reverse=True)[:n]

    recommended_books = [item[1] for item in top_n]

    return recommended_books
```
